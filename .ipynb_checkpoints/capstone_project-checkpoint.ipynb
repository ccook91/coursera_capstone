## Import packages and set options

```python
import numpy as np

import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

from geopy.geocoders import Nominatim
import folium

import requests

import urllib.request
import bs4 as bs

import seaborn as sns; sns.set()
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import DBSCAN
```

## Data Collection

### Gather Austin neighboorhoods and coordinates by zip code

```python
zips = pd.read_csv('texas_zips.csv')
austin_zips = zips[zips["City"] == "Austin"].reset_index(drop=True)
austin_zips.columns = ['zip_code', 'city', 'lat', 'lng']
austin_zips.head()
```

### Get population data by zip code

```python
pop_link = "http://zipatlas.com/us/tx/austin/zip-code-comparison/population-density.htm"
pop_source = urllib.request.urlopen(pop_link).read()

pop_html = bs.BeautifulSoup(pop_source, 'html.parser').find_all('table')[4].find_all('table')[7].find_next('table')
pop_df = pd.read_html(str(pop_html), header=0)[0]
pop_df.head()
```

```python
pop_df = pop_df[['Zip Code', 'Population', 'People / Sq. Mile']]
pop_df.columns = ['zip_code', 'pop', 'pop_density_sqmile']
pop_df.head()
```

### Get avg income data by zip code

```python
inc_link = "http://zipatlas.com/us/tx/austin/zip-code-comparison/average-income-per-person.htm"
inc_source = urllib.request.urlopen(inc_link).read()

inc_html = bs.BeautifulSoup(inc_source, 'html.parser').find_all('table')[4].find_all('table')[7].find_next('table')
inc_df = pd.read_html(str(inc_html), header=0)[0]
inc_df.head()
```

```python
inc_df = inc_df[['Zip Code', 'Avg. Income/Person']]
inc_df.columns = ['zip_code', 'avg_income']
inc_df['avg_income'] = inc_df['avg_income'].str[1:].str.replace(',','').astype('float64')
inc_df.head()
```

```python
austin = austin_zips.merge(pop_df, how='outer').merge(inc_df, how='left')
austin.head()
```

```python
austin = austin[pd.isnull(austin['pop'])==False]
austin.head()
```

```python
austin.dtypes
```

### Plot map of Austin and zip codes

```python
address = 'Austin, TX'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of {} are {}, {}.'.format(address, latitude, longitude))
```

```python
map_zips = folium.Map(location=[latitude, longitude], zoom_start=10)

for lat, lng, zip_code, pop, density, inc in zip(austin['lat'], austin['lng'], austin['zip_code'], austin['pop'], austin['pop_density_sqmile'], austin['avg_income']):
    label = '{} --\n  population: {}\n  population density: {}\n  avg income: {}'.format(zip_code, pop, density, inc)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_zips)  

map_zips
```
